```python
def log_loss(y_true, y_predicted):
    epsilon=1e-15
    y_=[max(i,epsilon) for i in y_predicted]
    y_=[min(i,1-epsilon) for i in y_]
    y_= np.array(y_)
    return -np.mean(y_true*np.log(y_) +(1-y_true)* np.log(1-y_))
```

```python

X = data.drop(['Timestamp', 'Clicked on Ad', 'Ad Topic Line', 'Country', 'City','Area Income','Daily Internet Usage','Male'], axis=1)
y = data['Clicked on Ad']
X['Age']=X['Age']/100
X['Daily Time Spent on Site']=X['Daily Time Spent on Site']/100
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)

```
